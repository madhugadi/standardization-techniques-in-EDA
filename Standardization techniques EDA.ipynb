###  Dealing with Outliers 

```python
 market=pd.read_csv('market_data.csv')
```

```python
market.head()
```

```python
market['Item_Weight'].describe()
```

```python
mark=market['Item_Visibility'].quantile([0.25,0.75])
mark
```

```python
iqr=(0.094585)-(0.026989)
iqr
```

```python
mark=market['Item_Visibility']
```

```python
ub=(0.094585)+((1.5)*(iqr))
print(ub)
lb=(0.026989)-((1.5)*(iqr))
print(lb)
```

```python
ls=[]
for i in mark:
    if i>=ub or i<=lb:
        ls.append(i)
print(ls)
len(ls)
        
        
        

        

        
        

        
```

### Capping

```python
market['Item_Visibility']=market['Item_Visibility'].replace(market[market['Item_Visibility']>ub]['Item_Visibility'],
                                                            market['Item_Visibility'].quantile(0.99))
```

```python
market[market['Item_Visibility']>ub]['Item_Visibility']
```

```python
market['Item_Visibility']=market['Item_Visibility'].replace(market[market['Item_Visibility']<lb]['Item_Visibility'],
                                                            market['Item_Visibility'].quantile(0.01))
```
